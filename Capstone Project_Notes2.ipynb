```python
def detect_outliers_iqr(df_cleaned):
    # Ensure all columns have numeric data types
    df_cleaned = df_cleaned.select_dtypes(include=[np.number])
    
    Q1 = df_cleaned.quantile(0.25)
    Q3 = df_cleaned.quantile(0.75)
    IQR = Q3 - Q1
    
    # Define outliers using IQR method
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Check for outliers in each column
    outliers = ((df_cleaned < lower_bound) | (df_cleaned > upper_bound)).any(axis=1)
    
    return outliers
```

```python
import numpy as np
# Drop non-numerical columns
numerical_df = df_cleaned.select_dtypes(include=[np.number])

# Calculate quartiles and IQR
Q1 = numerical_df.quantile(0.25)
Q3 = numerical_df.quantile(0.75)
IQR = Q3 - Q1

# Define outliers using IQR method
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_iqr = ((numerical_df < lower_bound) | (numerical_df > upper_bound)).any(axis=1)

# Filter DataFrame to show only rows with outliers
outliers_df = df_cleaned[outliers_iqr]
print("Outliers detected using IQR method:")
print(outliers_df)
```
